```python
import pandas as pd
import numpy as np

data = pd.read_csv("brain_stroke.csv")
data
```

```python
data.isnull().sum()
```

```python
print(data["gender"].value_counts())
print(data["ever_married"].value_counts())
print(data["work_type"].value_counts())
print(data["Residence_type"].value_counts())
print(data["smoking_status"].value_counts())
```

```python
cleanupGender = {"gender":{"Male":0, "Female":1}}
print(cleanupGender)

data = data.replace(cleanupGender)
data["gender"]
```

```python
cleanupMarried = {"ever_married":{"No":0, "Yes":1}}
print(cleanupMarried)

data = data.replace(cleanupMarried)
data["ever_married"]
```

```python
cleanupWork = {"work_type":{"Private":0, "Self-employed":1, "children":2, "Govt_job":3}}
print(cleanupWork)

data = data.replace(cleanupWork)
data["work_type"]
```

```python
# Urban    2532
# Rural    2449
# Name: Residence_type, dtype: int64
# never smoked       1838
# Unknown            1500
# formerly smoked     867
# smokes              776
# Name: smoking_status, dtype: int64

cleanupResidence = {"Residence_type":{"Urban":0, "Rural":1}}
print(cleanupResidence)

data = data.replace(cleanupResidence)
data["Residence_type"]
```

```python
cleanupSmoke = {"smoking_status":{"never smoked":0, "Unknown":1, "formerly smoked":2, "smokes":3}}
print(cleanupSmoke)

data = data.replace(cleanupSmoke)
data["work_type"]
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(data)
print(scaled)

from sklearn import model_selection
X = scaled[:,0:10]
Y = scaled [:,10]
seed = 7
validation_size = 0.3
X_train, X_test, Y_train, Y_test =model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
X_train
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Test options and evaluation metric
seed = 7
#the seed ensures we have the same sequence of random numbers. 
#The random numbers ensure we have a random split of the data into the k folds.
scoring = 'accuracy'
# Spot Check Algorithms
models = []
models.append(('LR',  LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier(n_neighbors=5)))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))

print(models)
```

```python
results=[]
names=[]
#putting a loop for testing various models
for name, model in models:
    kfold=model_selection.KFold(n_splits=15, random_state=seed,shuffle=True)
    cv_results = model_selection.cross_val_score(model,X_train,Y_train, cv=kfold)
    results.append(cv_results)
    names.append(name)
    msg = "%s : %f (%f)"%(name,cv_results.mean(),cv_results.std())
    print(msg)
    
print(results)
```

```python
from sklearn.svm import SVC
SVM = SVC()

kfold=model_selection.KFold(n_splits=15, random_state=seed,shuffle=True)
cv_results = model_selection.cross_val_score(SVM,X_train,Y_train, cv=kfold)
cv_results.mean()*100
```
